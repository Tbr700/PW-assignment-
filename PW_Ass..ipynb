# What is NumPy, and why is it widely used in Python?

# How does broadcasting work in NumPy
Broadcasting in NumPy refers to the ability of performing operations on arrays with different shapes by automatically expanding the smaller array's shape to match the larger array's shape. This is useful when performing arithmetic operations or applying functions to arrays of different dimensions.

# What is a heatmap, and when should it be used?
A heatmap is a two-dimensional data visualization technique that uses colors to represent numerical values in a matrix or grid format. Each cell’s color intensity or hue corresponds to the magnitude of the data point it represents. Typically, warmer colors (like reds) indicate higher values and cooler colors (like blues) indicate lower values.

# What does the term “vectorized operation” mean in NumPy?
A vectorized operation in NumPy refers to performing mathematical operations on entire arrays (vectors) at once without using explicit loops in Python code. Instead of iterating over each element in an array, NumPy applies the operation simultaneously to all elements using fast, optimized low-level implementations (usually in C).

# Why is NumPy’s array slicing faster than Python’s list slicing?
NumPy's array slicing is faster than Python's list slicing because:
Memory layout: NumPy arrays store data in contiguous blocks of memory, allowing for efficient access and manipulation.
Vectorized operations: NumPy operations are optimized for arrays, reducing overhead and improving performance.
Native code: NumPy's core operations are implemented in C, which is faster than Python.
Less memory allocation:NumPy arrays often return views, reducing memory allocation and copying.
