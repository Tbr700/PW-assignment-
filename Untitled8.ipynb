```python
import pandas as pd

df = pd.read_csv('/content/BIKE DETAILS.csv')
display(df.head(10))
```

```python
missing_values = df.isnull().sum()
display(missing_values)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20, 5))
sns.histplot(df['selling_price'], kde=True, bins=30)
plt.title('Distribution of Selling Prices')
plt.xlabel('Selling Price')
plt.ylabel('Frequency')
plt.show()
```

```python
avg_price_by_seller_type = df.groupby('seller_type')['selling_price'].mean().reset_index()

plt.figure(figsize=(15, 2))
sns.barplot(x='seller_type', y='selling_price', data=avg_price_by_seller_type)
plt.title('Average Selling Price by Seller Type')
plt.xlabel('Seller Type')
plt.ylabel('Average Selling Price')
plt.show()
```

```python
avg_km_by_owner_type = df.groupby('owner')['km_driven'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='owner', y='km_driven', data=avg_km_by_owner_type)
plt.title('Average Kilometers Driven by Ownership Type')
plt.xlabel('Ownership Type')
plt.ylabel('Average Kilometers Driven')
plt.show()
```

```python
print("Summary statistics of km_driven before outlier removal:")
display(df['km_driven'].describe())
Q1 = df['km_driven'].quantile(0.25)
Q3 = df['km_driven'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_no_outliers = df[(df['km_driven'] >= lower_bound) & (df['km_driven'] <= upper_bound)].copy()

print("\nSummary statistics of km_driven after outlier removal:")
display(df_no_outliers['km_driven'].describe())
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x='year', y='selling_price', data=df)
plt.title('Relationship between Year and Selling Price')
plt.xlabel('Year')
plt.ylabel('Selling Price')
plt.show()
```

```python
df_encoded = pd.get_dummies(df, columns=['seller_type'], drop_first=True)
display(df_encoded.head())
```

```python
plt.figure(figsize=(8, 6))
correlation_matrix = df_encoded.select_dtypes(include=['number']).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numeric Columns')
plt.show()
```

## Analysis Summary Report

Based on the analysis performed:

*   **Factors affecting selling price:** The correlation heatmap indicates that `ex_showroom_price` has the strongest positive correlation with `selling_price` (approximately 0.92), suggesting that the original price of the bike is a major factor in its resale value. `Year` also shows a positive correlation (approximately 0.40), implying that newer bikes tend to have higher selling prices. `km_driven` has a negative correlation (approximately -0.21), indicating that bikes with more kilometers driven generally have lower selling prices.
*   **Data Cleaning and Feature Engineering:**
    *   We identified that the `ex_showroom_price` column had a significant number of missing values (435 out of 1061 rows). We did not impute or remove these missing values in this analysis but noted their presence.
    *   We used the IQR method to detect and remove outliers from the `km_driven` column. This resulted in the removal of some high-kilometer entries, leading to a decrease in the mean and standard deviation of the `km_driven` distribution.
    *   We performed one-hot encoding on the `seller_type` column to convert it into a numerical format for potential use in modeling. The `seller_type_Individual` column was created to represent whether the seller was an individual.

This analysis provides initial insights into the factors influencing bike selling prices and highlights areas for further investigation and potential modeling.
